```python
imdb.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer(ngram_range=(1, 2), max_features=1000)
tf_idf.fit(imdb['clean_text'])
x_values = tf_idf.transform(imdb['clean_text'])  # matrix
y_values = imdb['sentiment']

from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x_values, y_values, test_size=0.2)
```

```python
from sklearn import metrics

def evaluation(y_pred, y_test):
    Accuracy = metrics.accuracy_score(y_pred, y_test)
    Recall = metrics.recall_score(y_pred=y_pred, y_true=y_test, pos_label='positive', average='micro')
    Precision = metrics.precision_score(y_pred=y_pred, y_true=y_test, pos_label='positive', average='micro')
    F1 = 2 * (Precision * Recall) / (Precision + Recall)

    print("Accuracy: ", Accuracy.round(4))
    print("Recall:", Recall.round(4))
    print("Precision:", Precision.round(4))
    print("F-1 score:", F1.round(4))
```

```python
from sklearn import naive_bayes
from sklearn.model_selection import cross_val_score, KFold
naive_bayes = naive_bayes.MultinomialNB()
naive_bayes.fit(x_train,y_train)
y_pred_valid = naive_bayes.predict(x_valid)
evaluation(y_pred_valid, y_valid)
print("Cross Validation Score:", cross_val_score(naive_bayes, x_valid, y_valid, cv= KFold(5, shuffle=True, random_state = 22)))
```

```python
from sklearn import svm
svm_implement = svm.SVC()
svm_implement.fit(x_train, y_train)
y_pred = svm_implement.predict(x_valid)
evaluation(y_pred, y_valid)
from sklearn.model_selection import cross_val_score, KFold
print("Cross Validation Score:", cross_val_score(svm_implement, x_valid, y_valid, cv=KFold(5, shuffle=True, random_state=22)).mean().round(4))
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader, RandomSampler, SequentialSampler, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

# Load the model
model_name = "dslim/bert-base-NER"
tokenizer = BertTokenizer.from_pretrained(model_name)
m = BertForSequenceClassification.from_pretrained(model_name)

# Load the dataset
df = pd.read_csv('/content/imdb_review_sentiment_analysis (1).csv')
# Assuming the CSV has 'review' and 'sentiment' columns
# 'sentiment' should be 0 for negative and 1 for positive
# Preprocessing
def preprocess(texts, labels, tokenizer, max_len=512):
    input_ids, attention_masks, labels_out = [], [], []

    for text, label in zip(texts, labels):
        label = 0 if label == 'negative' else 1

        encoded_dict = tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        input_ids.append(encoded_dict['input_ids'])
        attention_masks.append(encoded_dict['attention_mask'])
        labels_out.append(label)

    input_ids = torch.cat(input_ids, dim=0)
    attention_masks = torch.cat(attention_masks, dim=0)
    labels_out = torch.tensor(labels_out)

    return input_ids, attention_masks, labels_out

train_texts, test_texts, train_labels, test_labels = train_test_split(df['clean_text'], df['sentiment'], test_size=0.3)
train_inputs, train_masks, train_labels = preprocess(train_texts, train_labels, tokenizer)
test_inputs, test_masks, test_labels = preprocess(test_texts, test_labels, tokenizer)

# Create dataloaders
batch_size = 16
train_data = TensorDataset(train_inputs, train_masks, train_labels)
train_sampler = RandomSampler(train_data)
train_dataloader = DataLoader(train_data, sampler=train_sampler, batch_size=batch_size)

test_data = TensorDataset(test_inputs, test_masks, test_labels)
test_sampler = SequentialSampler(test_data)
test_dataloader = DataLoader(test_data, sampler=test_sampler, batch_size=batch_size)


def evaluate(dataloader):
    m.eval()
    predictions, true_labels = [], []

    for batch in dataloader:
        b_input_ids, b_input_mask, b_labels = batch
        with torch.no_grad():
            outputs = m(b_input_ids, token_type_ids=None, attention_mask=b_input_mask)

        logits = outputs[0]
        logits = logits.detach().cpu().numpy()
        label_ids = b_labels.to('cpu').numpy()

        predictions.extend(logits)
        true_labels.extend(label_ids)

    # Convert lists to numpy arrays before flattening
    predictions = np.array(predictions)
    true_labels = np.array(true_labels)

    pred_flat = np.argmax(predictions, axis=1).flatten()
    labels_flat = true_labels.flatten()
    return accuracy_score(labels_flat, pred_flat)

# Evaluate the model
accuracy = evaluate(test_dataloader)
print("Accuracy:", accuracy)
```
